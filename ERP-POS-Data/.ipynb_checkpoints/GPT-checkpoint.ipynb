```python
# Script completo - copiar y ejecutar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import os

# ---------- AJUSTA la ruta del CSV si hace falta ----------
csv_path = "Sales-CSV970-1093.csv"   # <-- pon aquí la ruta correcta si no está en el mismo directorio

# 1) Cargar datos (detecta separador ';' si es necesario)
try:
    df = pd.read_csv(csv_path, sep=';', encoding='utf-8', low_memory=False)
except Exception:
    df = pd.read_csv(csv_path, encoding='utf-8', low_memory=False)

print("Datos cargados. Shape:", df.shape)
print("Columnas:", list(df.columns)[:40])

# 2) Verificar que exista la columna objetivo
target = "TotalVentaNeta"
if target not in df.columns:
    raise ValueError(f"No se encontró la columna objetivo '{target}'. Revisa el nombre exacto en el CSV.")

# 3) Trabajar sólo con columnas numéricas (y crear lag)
df_num = df.select_dtypes(include=[np.number]).copy()

# Crear lag 1 de TotalVentaNeta (autoregresivo) — útil para dinámica discreta
df_num['TotalVentaNeta_lag1'] = df_num[target].shift(1)

# 4) Imputar NaNs con mediana (por columna numérica)
df_num = df_num.fillna(df_num.median())

# 5) Seleccionar features: todas numéricas menos target, luego elegir top correlacionadas
all_numeric = [c for c in df_num.columns if c != target]
corrs = df_num[all_numeric + [target]].corr()[target].abs().sort_values(ascending=False)

# Elegir top k (incluye lag1 si es relevante)
k = 8
top_features = corrs.index[:k].tolist()
# Asegurar que TotalVentaNeta_lag1 esté presente (prioritario para dinámica)
if 'TotalVentaNeta_lag1' not in top_features:
    top_features = ['TotalVentaNeta_lag1'] + [f for f in top_features if f!='TotalVentaNeta_lag1'][:k-1]

print("\nTop features seleccionadas (por correlación absoluta):")
for i,f in enumerate(top_features,1):
    print(f"{i}. {f} (|corr|={corrs.get(f):.4f})")

# 6) Preparar X, y arrays (preservando orden temporal)
X = df_num[top_features].values
y = df_num[target].values

n = len(df_num)
split = int(0.9 * n)  # 90% entrenamiento
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"\nObservaciones: total={n}, train={len(X_train)}, test={len(X_test)}")

# 7) Escalado para el modelo lineal
scaler = StandardScaler().fit(X_train)
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)

# 8) Entrenar modelos
lr = LinearRegression().fit(X_train_s, y_train)
rf = RandomForestRegressor(n_estimators=300, random_state=42).fit(X_train, y_train)

# 9) Predicciones
y_pred_lr = lr.predict(X_test_s)
y_pred_rf = rf.predict(X_test)

# 10) Métricas
def metrics(y_true, y_pred):
    return {
        "R2": r2_score(y_true, y_pred),
        "MSE": mean_squared_error(y_true, y_pred),
        "MAE": mean_absolute_error(y_true, y_pred)
    }

metrics_lr = metrics(y_test, y_pred_lr)
metrics_rf = metrics(y_test, y_pred_rf)

print("\nMétricas (conjunto prueba):")
print("Linear Regression -> R2: {:.4f}, MSE: {:.4f}, MAE: {:.4f}".format(metrics_lr['R2'], metrics_lr['MSE'], metrics_lr['MAE']))
print("Random Forest    -> R2: {:.4f}, MSE: {:.4f}, MAE: {:.4f}".format(metrics_rf['R2'], metrics_rf['MSE'], metrics_rf['MAE']))

# 11) Matriz de correlación (mostrar heatmap)
plt.figure(figsize=(10,8))
corr_matrix = df_num.corr()
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Matriz de correlación (numéricas)")
plt.tight_layout()
plt.show()

# 12) Gráficas: TotalVentaNeta real (negro) y simulación (RF en rojo)
plt.figure(figsize=(12,5))
plt.plot(range(len(y_test)), y_test, color='black', label='TotalVentaNeta (real)')
plt.plot(range(len(y_test)), y_pred_rf, color='red', linestyle='--', label='Predicción RF (simulación)')
plt.plot(range(len(y_test)), y_pred_lr, color='blue', linestyle=':', label='Predicción LR')
plt.xlabel("Índice (prueba, orden temporal)")
plt.ylabel("TotalVentaNeta")
plt.title("TotalVentaNeta real (negro) vs simulaciones (RF en rojo, LR en azul)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 13) Presentar modelo lineal en escala original (coeficientes interpretables)
coef_scaled = lr.coef_               # coeficientes sobre features escaladas
scale = scaler.scale_
mean = scaler.mean_

# transformar coeficientes a escala original:
coef_original = coef_scaled / scale
intercept_original = lr.intercept_ - np.sum(coef_scaled * mean / scale)

# Preparar y mostrar ecuación
eq_terms = [f"({coef_original[i]:+.6f})*{top_features[i]}" for i in range(len(top_features))]
equation = f"{target} = {intercept_original:+.6f} " + " ".join(eq_terms)
print("\nEcuación del modelo lineal (coeficientes en escala original):")
print(equation)

# 14) Mostrar tablas útiles: coeficientes y feature importances
coef_df = pd.DataFrame({
    "feature": top_features,
    "coef_scaled": coef_scaled,
    "coef_original": coef_original
}).round(6)

rf_imp = pd.Series(rf.feature_importances_, index=top_features).sort_values(ascending=False).round(6)

print("\nCoeficientes (Linear Regression):")
print(coef_df.to_string(index=False))

print("\nImportancia de features (Random Forest):")
print(rf_imp.to_string())

# 15) Guardar resultados en archivos (opcional)
out_dir = "./"  # o "/mnt/data/" si corres en entorno con ese path
os.makedirs(out_dir, exist_ok=True)
coef_df.to_csv(os.path.join(out_dir, "lr_coefficients.csv"), index=False)
pd.DataFrame([metrics_lr, metrics_rf], index=["LinearRegression","RandomForest"]).to_csv(os.path.join(out_dir,"model_metrics.csv"))
rf_imp.to_csv(os.path.join(out_dir, "rf_importances.csv"))
corr_matrix.to_csv(os.path.join(out_dir, "correlation_matrix.csv"))
print(f"\nArchivos guardados en {os.path.abspath(out_dir)}: lr_coefficients.csv, model_metrics.csv, rf_importances.csv, correlation_matrix.csv")
```
