```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer el archivo CSV
df = pd.read_csv('../ERP-POS-Data/Sales-CSV970-1093.csv', delimiter=';')
df = df.drop("EsFechaEspecial", axis=1)
# Eliminamos los caracteres no deseados y convertimos a tipo numérico
# display(df)
```

```python
# 1️⃣ Calcular la matriz de correlación
correlation_matrix = df.corr()
```

```python
print("Matriz de correlación:")
#print(correlation_matrix)
```

```python
# 2️⃣ Visualizar con un heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de correlación")
plt.show()
```

```python
plt.boxplot(df['TotalVentaNeta'].head(970))
plt.title("Total venta neta")
plt.ylabel("Ventas")
plt.show()

#La caja (rectángulo) representa el rango intercuartílico (del 25% al 75% de los datos).
#La línea dentro de la caja es la mediana de las ventas netas.
#Los bigotes se extienden hasta un rango definido (1.5 veces el rango intercuartílico).
#Los círculos son outliers (valores atípicos), en este caso hay muchos valores extremos de ventas muy superiores al rango típico.
```

```python
plt.boxplot(df['UnidadesKit'].head(970))
plt.title("UnidadesKit")
plt.ylabel("Ventas")
plt.show()

#La mediana está alrededor de 40 unidades.
#La caja indica que el 50 % central de los datos está entre aproximadamente 25 y 50 unidades.
#Los bigotes llegan hasta cerca de 90 unidades.
#Existen muchos outliers por encima de ese rango, algunos muy altos (hasta más de 250 unidades).
```

```python
plt.boxplot(df['VentasCENTROORIENTE'].head(970))
plt.title("VentasCENTROORIENTE")
plt.ylabel("Ventas")
plt.show()

#Mediana: ~1,000,000 (línea central de la caja)
#Rango intercuartílico (IQR): Entre ~700,000 y ~1,500,000
#Múltiples outliers superiores: Varios puntos atípicos entre 2-9 millones
#Ventas típicas concentradas: El 75% de las ventas están por debajo de 1.5 millones, indicando un comportamiento base relativamente estable
```

```python
plt.boxplot(df['VentasOCCIDENTE'].head(970))
plt.title("VentasOCCIDENTE")
plt.ylabel("Ventas")
plt.show()

#Mediana: ~600,000 (ligeramente menor que Centro Oriente)
#IQR: Entre ~400,000 y ~900,000
#Outliers superiores: Múltiples valores atípicos entre 2-7 millones
#La caja está posicionada más abajo que Centro Oriente, indicando que las ventas típicas son menores en esta región (~40% menos en la mediana)
```

```python
plt.boxplot(df['VentasNORTE'].head(970))
plt.title("VentasNORTE")
plt.ylabel("Ventas")
plt.show()

#Mediana: ~250,000 (significativamente menor que las otras dos regiones)
#IQR: Entre ~100,000 y ~600,000 (más compacto)
#Outliers: Concentrados entre 1.5-3.2 millones, pero con una diferencia clave
#Volumen base mucho menor: La mediana es apenas el 25-40% de las otras regiones
```
